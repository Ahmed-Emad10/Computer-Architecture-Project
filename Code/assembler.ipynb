```python
import re
from enum import Enum

class registers(Enum):
    R0 = "000"
    R1 = "001"
    R2 = "010"
    R3 = "011"
    R4 = "100"
    R5 = "101"
    R6 = "110"
    R7 = "111"
```

```python
# should I make everything 32 bits or keep 16 bits? and imm is hex? and I kept MOV like the doc..
f = open("assembly.txt","r")
f2 = open("instruction.txt","w")
for x in f:
    arr = re.split(r'[(), \s]\s*', x)
    if(arr[0].lower()=="nop"):
        ins = "00000000000000000000000000000000"
    elif(arr[0].lower() == "hlt"):
        ins = "00001000000000000000000000000000"
    elif(arr[0].lower() == "setc"):
        ins = "00010000000000000000000000000000"
    elif(arr[0].lower() == "not"):
        ins = "00011" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "inc"):
        ins = "00100" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "out"):
        ins = "00101" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "in"):
        ins = "00110" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "mov"):
        ins = "01000" + registers[arr[1]].value + registers[arr[2]].value + "000000000000000000000"
    elif(arr[0].lower() == "add"):
        ins = "01001" + registers[arr[2]].value + registers[arr[3]].value + registers[arr[1]].value + "000000000000000000"
    elif(arr[0].lower() == "sub"):
        ins = "01010" + registers[arr[2]].value + registers[arr[3]].value + registers[arr[1]].value + "000000000000000000"
    elif(arr[0].lower() == "and"):
        ins = "01011" + registers[arr[2]].value + registers[arr[3]].value + registers[arr[1]].value + "000000000000000000"
    elif(arr[0].lower() == "iadd"):
        ins = "01100" + registers[arr[2]].value + registers[arr[1]].value + "00000" + arr[3]
    elif(arr[0].lower() == "push"):
        ins = "10000" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "pop"):
        ins = "10001" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "ldm"):
        ins = "10010" + registers[arr[1]].value + "00000000" + arr[2]
    elif(arr[0].lower() == "ldd"):
        ins = "10011" + registers[arr[3]].value + registers[arr[1]].value + "00000" + arr[2]
    elif(arr[0].lower() == "std"):
        ins = "10100" + registers[arr[3]].value + registers[arr[1]].value + "00000" + arr[2]
    elif(arr[0].lower() == "jz"):
        ins = "11000" + registers[arr[1]].value + "000000000000000000000000" 
    elif(arr[0].lower() == "jn"):
        ins = "11001" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "jc"):
        ins = "11010" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "jmp"):
        ins = "11011" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "call"):
        ins = "11100" + registers[arr[1]].value + "000000000000000000000000"
    elif(arr[0].lower() == "ret"):
        ins = "11101" + "000000000000000000000000000" 
    elif(arr[0].lower() == "int"):
        ins = "11110" + arr[1] + "0000000000000000000000000"
    elif(arr[0].lower() == "rti"):
        ins = "11111" + "000000000000000000000000000" 
    f2.write(ins+'\n')
f.close()
f2.close()
```
